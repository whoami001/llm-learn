````python
import os
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

# åŠ è½½ç¯å¢ƒå˜é‡
_ = load_dotenv(find_dotenv())

# åˆå§‹åŒ– DeepSeek å®¢æˆ·ç«¯
client = OpenAI(
    api_key=os.getenv('DEEPSEEK_API_KEY'),
    base_url="https://api.deepseek.com"  # DeepSeek API ç«¯ç‚¹
)

def get_completion(prompt, model="deepseek-chat"):
    """
    è°ƒç”¨ DeepSeek èŠå¤©æ¨¡å‹è·å–å›å¤
    
    å‚æ•°:
    prompt: ç”¨æˆ·è¾“å…¥çš„æç¤ºè¯
    model: æ¨¡å‹åç§°ï¼Œé»˜è®¤ä¸º deepseek-chat
          å…¶ä»–å¯é€‰: deepseek-coder, deepseek-reasoner ç­‰
    
    è¿”å›:
    æ¨¡å‹ç”Ÿæˆçš„å›å¤å†…å®¹
    """
    messages = [{"role": "user", "content": prompt}]
    
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,  # æ¸©åº¦å‚æ•°ï¼Œæ§åˆ¶éšæœºæ€§
        # å…¶ä»–å¯é€‰å‚æ•°:
        # max_tokens=2048,  # æœ€å¤§ç”Ÿæˆ token æ•°
        # stream=False,     # æ˜¯å¦æµå¼è¾“å‡º
    )
    
    return response.choices[0].message.content

# ä½¿ç”¨ç¤ºä¾‹
if __name__ == "__main__":
    # æµ‹è¯•è°ƒç”¨
    prompt = "è¯·ä»‹ç»ä¸€ä¸‹ä½ è‡ªå·±"
    result = get_completion(prompt)
    print(result)
````

````python
def get_completion_from_messages(messages,temperature=0):
    '''
    å°è£…ä¸€ä¸ªæ”¯æŒæ›´å¤šå‚æ•°çš„è‡ªå®šä¹‰è®¿é—® deepseek-chatæ¨¡å‹çš„å‡½æ•°

    å‚æ•°: 
    messages: è¿™æ˜¯ä¸€ä¸ªæ¶ˆæ¯åˆ—è¡¨ï¼Œæ¯ä¸ªæ¶ˆæ¯éƒ½æ˜¯ä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å« role(è§’è‰²ï¼‰å’Œ content(å†…å®¹)ã€‚è§’è‰²å¯ä»¥æ˜¯'system'ã€'user' æˆ– 'assistantâ€™ï¼Œå†…å®¹æ˜¯è§’è‰²çš„æ¶ˆæ¯ã€‚
    model: è°ƒç”¨çš„æ¨¡å‹ï¼Œé»˜è®¤ä¸º gpt-3.5-turbo(ChatGPT)ï¼Œæœ‰å†…æµ‹èµ„æ ¼çš„ç”¨æˆ·å¯ä»¥é€‰æ‹© gpt-4
    temperature: è¿™å†³å®šæ¨¡å‹è¾“å‡ºçš„éšæœºç¨‹åº¦ï¼Œé»˜è®¤ä¸º0ï¼Œè¡¨ç¤ºè¾“å‡ºå°†éå¸¸ç¡®å®šã€‚å¢åŠ æ¸©åº¦ä¼šä½¿è¾“å‡ºæ›´éšæœºã€‚
    max_tokens: è¿™å†³å®šæ¨¡å‹è¾“å‡ºçš„æœ€å¤§çš„ token æ•°ã€‚
    '''
   
    model="deepseek-chat"
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,  # æ¸©åº¦å‚æ•°ï¼Œæ§åˆ¶éšæœºæ€§
        # å…¶ä»–å¯é€‰å‚æ•°:
        # max_tokens=2048,  # æœ€å¤§ç”Ÿæˆ token æ•°
        # stream=False,     # æ˜¯å¦æµå¼è¾“å‡º
    )
    
    return response.choices[0].message.content
````

````python
def get_completion_and_token_count(messages, 
                                   model="deepseek-chat", 
                                   temperature=0, 
                                   max_tokens=500):
    """
    ä½¿ç”¨ OpenAI çš„ deepseek-chat æ¨¡å‹ç”ŸæˆèŠå¤©å›å¤ï¼Œå¹¶è¿”å›ç”Ÿæˆçš„å›å¤å†…å®¹ä»¥åŠä½¿ç”¨çš„ token æ•°é‡ã€‚
    å‚æ•°:
    messages: èŠå¤©æ¶ˆæ¯åˆ—è¡¨ã€‚
    model: ä½¿ç”¨çš„æ¨¡å‹åç§°ã€‚é»˜è®¤ä¸º"deepseek-chat"ã€‚
    temperature: æ§åˆ¶ç”Ÿæˆå›å¤çš„éšæœºæ€§ã€‚å€¼è¶Šå¤§ï¼Œç”Ÿæˆçš„å›å¤è¶Šéšæœºã€‚é»˜è®¤ä¸º 0ã€‚
    max_tokens: ç”Ÿæˆå›å¤çš„æœ€å¤§ token æ•°é‡ã€‚é»˜è®¤ä¸º 500ã€‚
    è¿”å›:
    content: ç”Ÿæˆçš„å›å¤å†…å®¹ã€‚
    token_dict: åŒ…å«'prompt_tokens'ã€'completion_tokens'å’Œ'total_tokens'çš„å­—å…¸ï¼Œåˆ†åˆ«è¡¨ç¤ºæç¤ºçš„ token æ•°é‡ã€ç”Ÿæˆçš„å›å¤çš„ token æ•°é‡å’Œæ€»çš„ token æ•°é‡ã€‚
    """
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, 
        max_tokens=max_tokens,
    )
    # âœ… æ­£ç¡®å†™æ³•
    content = response.choices[0].message.content
    token_dict = {
        'prompt_tokens': response.usage.prompt_tokens,  # ä½¿ç”¨å±æ€§è®¿é—®
        'completion_tokens': response.usage.completion_tokens,
        'total_tokens': response.usage.total_tokens,
    }
    return content, token_dict
````

````python
def get_product_by_name(name):
    """
    æ ¹æ®äº§å“åç§°è·å–äº§å“

    å‚æ•°:
    name: äº§å“åç§°
    """
    return products.get(name, None)

def get_products_by_category(category):
    """
    æ ¹æ®ç±»åˆ«è·å–äº§å“

    å‚æ•°:
    category: äº§å“ç±»åˆ«
    """
    return [product for product in products.values() if product["ç±»åˆ«"] == category]
````

````python
def read_string_to_list(input_string):
    """
    å°†è¾“å…¥çš„å­—ç¬¦ä¸²è½¬æ¢ä¸º Python åˆ—è¡¨ã€‚

    å‚æ•°:
    input_string: è¾“å…¥çš„å­—ç¬¦ä¸²ï¼Œåº”ä¸ºæœ‰æ•ˆçš„ JSON æ ¼å¼ã€‚

    è¿”å›:
    list æˆ– None: å¦‚æœè¾“å…¥å­—ç¬¦ä¸²æœ‰æ•ˆï¼Œåˆ™è¿”å›å¯¹åº”çš„ Python åˆ—è¡¨ï¼Œå¦åˆ™è¿”å› Noneã€‚
    """
    if input_string is None:
        return None

    try:
        # å°†è¾“å…¥å­—ç¬¦ä¸²ä¸­çš„å•å¼•å·æ›¿æ¢ä¸ºåŒå¼•å·ï¼Œä»¥æ»¡è¶³ JSON æ ¼å¼çš„è¦æ±‚
        input_string = input_string.replace("'", "\"")  
        data = json.loads(input_string)
        return data
    except json.JSONDecodeError:
        print("Error: Invalid JSON string")
        return None   

category_and_product_list = read_string_to_list(category_and_product_response_1)
print(category_and_product_list)
````

````python
def generate_output_string(data_list):
    """
    æ ¹æ®è¾“å…¥çš„æ•°æ®åˆ—è¡¨ç”ŸæˆåŒ…å«äº§å“æˆ–ç±»åˆ«ä¿¡æ¯çš„å­—ç¬¦ä¸²ã€‚

    å‚æ•°:
    data_list: åŒ…å«å­—å…¸çš„åˆ—è¡¨ï¼Œæ¯ä¸ªå­—å…¸éƒ½åº”åŒ…å« "products" æˆ– "category" çš„é”®ã€‚

    è¿”å›:
    output_string: åŒ…å«äº§å“æˆ–ç±»åˆ«ä¿¡æ¯çš„å­—ç¬¦ä¸²ã€‚
    """
    output_string = ""
    if data_list is None:
        return output_string

    for data in data_list:
        try:
            if "products" in data and data["products"]:
                products_list = data["products"]
                for product_name in products_list:
                    product = get_product_by_name(product_name)
                    if product:
                        output_string += json.dumps(product, indent=4, ensure_ascii=False) + "\n"
                    else:
                        print(f"Error: Product '{product_name}' not found")
            elif "category" in data:
                category_name = data["category"]
                category_products = get_products_by_category(category_name)
                for product in category_products:
                    output_string += json.dumps(product, indent=4, ensure_ascii=False) + "\n"
            else:
                print("Error: Invalid object format")
        except Exception as e:
            print(f"Error: {e}")

    return output_string 

product_information_for_user_message_1 = generate_output_string(category_and_product_list)
print(product_information_for_user_message_1)
````

````python
# è°ƒç”¨ä¸­æ–‡ Prompt ç‰ˆæœ¬
def collect_messages_ch(debug=True):
    """
    ç”¨äºæ”¶é›†ç”¨æˆ·çš„è¾“å…¥å¹¶ç”ŸæˆåŠ©æ‰‹çš„å›ç­”

    å‚æ•°ï¼š
    debug: ç”¨äºè§‰å¾—æ˜¯å¦å¼€å¯è°ƒè¯•æ¨¡å¼
    """
    user_input = inp.value_input
    if debug: print(f"User Input = {user_input}")
    if user_input == "":
        return
    inp.value = ''
    global context
    # è°ƒç”¨ process_user_message å‡½æ•°
    #response, context = process_user_message(user_input, context, utils.get_products_and_category(),debug=True)
    response, context = process_user_message_ch(user_input, context, debug=False)
    # print(response)
    context.append({'role':'assistant', 'content':f"{response}"})
    panels.append(
        pn.Row('User:', pn.pane.Markdown(user_input, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600, styles={'background-color': '#F6F6F6'})))
 
    return pn.Column(*panels) # åŒ…å«äº†æ‰€æœ‰çš„å¯¹è¯ä¿¡æ¯
````

````python
import json
def eval_response_with_ideal(response,
                              ideal,
                              debug=False):
    """
    è¯„ä¼°å›å¤æ˜¯å¦ä¸ç†æƒ³ç­”æ¡ˆåŒ¹é…
    
    å‚æ•°ï¼š
    response: å›å¤çš„å†…å®¹
    ideal: ç†æƒ³çš„ç­”æ¡ˆ
    debug: æ˜¯å¦æ‰“å°è°ƒè¯•ä¿¡æ¯
    """
    if debug:
        print("å›å¤ï¼š")
        print(response)
    
    # json.loads() åªèƒ½è§£æåŒå¼•å·ï¼Œå› æ­¤æ­¤å¤„å°†å•å¼•å·æ›¿æ¢ä¸ºåŒå¼•å·
    json_like_str = response.replace("'",'"')
    
    # è§£æä¸ºä¸€ç³»åˆ—çš„å­—å…¸
    l_of_d = json.loads(json_like_str)
    
    # å½“å“åº”ä¸ºç©ºï¼Œå³æ²¡æœ‰æ‰¾åˆ°ä»»ä½•å•†å“æ—¶
    if l_of_d == [] and ideal == []:
        return 1
    
    # å¦å¤–ä¸€ç§å¼‚å¸¸æƒ…å†µæ˜¯ï¼Œæ ‡å‡†ç­”æ¡ˆæ•°é‡ä¸å›å¤ç­”æ¡ˆæ•°é‡ä¸åŒ¹é…
    elif l_of_d == [] or ideal == []:
        return 0
    
    # ç»Ÿè®¡æ­£ç¡®ç­”æ¡ˆæ•°é‡
    correct = 0    
    
    if debug:
        print("l_of_d is")
        print(l_of_d)

    # å¯¹æ¯ä¸€ä¸ªé—®ç­”å¯¹  
    for d in l_of_d:

        # è·å–äº§å“å’Œç›®å½•
        cat = d.get('category')
        prod_l = d.get('products')
        # æœ‰è·å–åˆ°äº§å“å’Œç›®å½•
        if cat and prod_l:
            # convert list to set for comparison
            prod_set = set(prod_l)
            # get ideal set of products
            ideal_cat = ideal.get(cat)
            if ideal_cat:
                prod_set_ideal = set(ideal.get(cat))
            else:
                if debug:
                    print(f"æ²¡æœ‰åœ¨æ ‡å‡†ç­”æ¡ˆä¸­æ‰¾åˆ°ç›®å½• {cat}")
                    print(f"æ ‡å‡†ç­”æ¡ˆ: {ideal}")
                continue
                
            if debug:
                print("äº§å“é›†åˆï¼š\n",prod_set)
                print()
                print("æ ‡å‡†ç­”æ¡ˆçš„äº§å“é›†åˆï¼š\n",prod_set_ideal)

            # æŸ¥æ‰¾åˆ°çš„äº§å“é›†åˆå’Œæ ‡å‡†çš„äº§å“é›†åˆä¸€è‡´
            if prod_set == prod_set_ideal:
                if debug:
                    print("æ­£ç¡®")
                correct +=1
            else:
                print("é”™è¯¯")
                print(f"äº§å“é›†åˆ: {prod_set}")
                print(f"æ ‡å‡†çš„äº§å“é›†åˆ: {prod_set_ideal}")
                if prod_set <= prod_set_ideal:
                    print("å›ç­”æ˜¯æ ‡å‡†ç­”æ¡ˆçš„ä¸€ä¸ªå­é›†")
                elif prod_set >= prod_set_ideal:
                    print("å›ç­”æ˜¯æ ‡å‡†ç­”æ¡ˆçš„ä¸€ä¸ªè¶…é›†")

    # è®¡ç®—æ­£ç¡®ç­”æ¡ˆæ•°
    pc_correct = correct / len(l_of_d)
        
    return pc_correct
````
