````python
import os
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

# 加载环境变量
_ = load_dotenv(find_dotenv())

# 初始化 DeepSeek 客户端
client = OpenAI(
    api_key=os.getenv('DEEPSEEK_API_KEY'),
    base_url="https://api.deepseek.com"  # DeepSeek API 端点
)

def get_completion(prompt, model="deepseek-chat"):
    """
    调用 DeepSeek 聊天模型获取回复
    
    参数:
    prompt: 用户输入的提示词
    model: 模型名称，默认为 deepseek-chat
          其他可选: deepseek-coder, deepseek-reasoner 等
    
    返回:
    模型生成的回复内容
    """
    messages = [{"role": "user", "content": prompt}]
    
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,  # 温度参数，控制随机性
        # 其他可选参数:
        # max_tokens=2048,  # 最大生成 token 数
        # stream=False,     # 是否流式输出
    )
    
    return response.choices[0].message.content

# 使用示例
if __name__ == "__main__":
    # 测试调用
    prompt = "请介绍一下你自己"
    result = get_completion(prompt)
    print(result)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

# 加载环境变量
_ = load_dotenv(find_dotenv())

# 初始化DeepSeek客户端
client = OpenAI(
    api_key=os.getenv('DEEPSEEK_API_KEY'),  # 修改为DeepSeek的API密钥
    base_url="https://api.deepseek.com"  # 修改API端点
)

def get_completion(prompt, model="deepseek-chat"):
    """
    调用DeepSeek聊天模型获取回复
    """
    messages = [{"role": "user", "content": prompt}]
    
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    
    return response.choices[0].message.content

def get_completion_from_messages(messages, model="deepseek-chat", temperature=0):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content
````
